# **4. FEATURE CREATION**

***Aggregate Curricular Units: You have columns for Curricular units for both semesters. Let's create aggregate features such as total credited units, total approved units, pass rate, and total grades.***

```python
# Total credited units (sum of 1st and 2nd semesters)
data['Total credited units'] = data['curricular_units_1st_sem_credited'] + data['curricular_units_2nd_sem_credited']

# Total approved units (sum of approved units across both semesters)
data['Total approved units'] = data['curricular_units_1st_sem_approved'] + data['curricular_units_2nd_sem_approved']

#Total evaluation units
data['Total evaluation units'] = data['curricular_units_1st_sem_evaluations'] + data['curricular_units_2nd_sem_evaluations']

#Total without evaluation units
data['Total without evaluation units'] = data['curricular_units_1st_sem_without_evaluations'] + data['curricular_units_2nd_sem_without_evaluations']

#Total enrolled units
data['Total enrolled units'] = data['curricular_units_1st_sem_enrolled'] + data['curricular_units_2nd_sem_enrolled']
```

```python
# Pass rate (approved units/enrolled units)
data['Pass rate'] = (data['Total approved units'] / 
                   (data['curricular_units_1st_sem_enrolled'] + data['curricular_units_2nd_sem_enrolled'])).fillna(0)
```

```python
# Total grades (average grade across all semesters)
data['Average grade'] = (data['curricular_units_1st_sem_grade'] + data['curricular_units_2nd_sem_grade']) / 2
```

```python
#Course Dropout Ratio
data['Dropout Ratio'] = (data['Total without evaluation units'] / data['Total enrolled units']).fillna(0)
                   
```

```python
#Improvement Rate Between Semesters
data['Improvement Rate'] = ((data['curricular_units_2nd_sem_grade']-data['curricular_units_1st_sem_grade'])/data['curricular_units_1st_sem_grade']).fillna(0)
```

```python
#Overall GPA (Grade Point Average)
data['Overall GPA'] = ((data['curricular_units_2nd_sem_grade']+data['curricular_units_1st_sem_grade'])/2).fillna(0)
```

 ***Bin Continuous Variables into Categorical Ones***

```python
age_bins = [0, 18, 25, 30, 40, 100]  
age_labels = ['Teenage', 'Young adult', 'Adult', 'Mid-age adult', 'Senior']  
binned_data = pd.cut(data['age_at_enrollment'], bins= age_bins, labels = age_labels)

data['life_stage'] = binned_data
data['life_stage'] = data['life_stage'].astype('category')
```

***Interaction Terms***

```python
# Create interaction features between 'Average grade' and 'Course'
data['AverageGrade_Course_interaction'] = data['Average grade'] * data['course'].factorize()[0]
```

***PolynomialFeatures***

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

# Select numerical columns to create polynomial features (ignore non-numeric columns)
numerical_cols = data.select_dtypes(include=[np.number]).columns

# Check for missing or infinite values in numerical columns
# Replace infinite values with NaN and then fill NaN values if necessary
data[numerical_cols] = data[numerical_cols].replace([np.inf, -np.inf], np.nan)
data[numerical_cols] = data[numerical_cols].fillna(0)  # You can choose a better strategy (e.g., median)

# Create polynomial features (degree 2)
poly = PolynomialFeatures(degree=2, include_bias=False)

# Generate polynomial features only for numerical columns
poly_features = poly.fit_transform(data[numerical_cols])

# Convert the polynomial features into a DataFrame
poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(numerical_cols))

# Concatenate the polynomial features with the original dataset
data_with_poly = pd.concat([data, poly_df], axis=1)

# Display the updated dataset with polynomial features
print(data_with_poly.head())
```

***B. Chi-square Test:***

```python
from scipy.stats import chi2_contingency

# Example: Chi-square test between 'target' and 'gender'
contingency_table = pd.crosstab(data['target'], data['gender'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2}, P-value: {p}")
```

***C. t-tests (Continuous Variables)***

```python

from scipy.stats import ttest_ind

dropouts = data[data['target'] == 'Dropout']['admission_grade']
non_dropouts = data[data['target'] != 'Dropout']['admission_grade']


t_stat, p_value = ttest_ind(dropouts, non_dropouts)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

*Insight: dropouts and non-dropouts differ significantly in terms of their admission grades, with dropouts typically having lower grades. The statistical significance of this difference is very high, suggesting that the relationship is unlikely to be due to random variation.*

# **9. DIMENSIONALITY REDUCTION**

***1. Principal Component Analysis (PCA):***

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Standardize the numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data.drop('target_encoded', axis=1))  # Exclude the target variable

# Apply PCA
pca = PCA(n_components=5)  # You can change the number of components
X_pca = pca.fit_transform(X_scaled)

# Plot the explained variance ratio to decide the number of components
explained_variance = pca.explained_variance_ratio_
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.show()

# Display the PCA-transformed data
pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
print(pca_df.head())
```

***Visualizing PCA Results:***

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data['target_encoded'], cmap='plasma')
plt.title('PCA: First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
