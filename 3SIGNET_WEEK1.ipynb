***IDENTIFY THE OUTLIERS***

```python
Q1 = numerical_df.quantile(0.25)
Q3 = numerical_df.quantile(0.75)
IQR = Q3 - Q1
```

```python
outlier_condition = (numerical_df < (Q1 - 1.5 * IQR)) | \
                    (numerical_df > (Q3 + 1.5 * IQR))
outliers = data[outlier_condition.any(axis=1)]
```

## **FEATURE ENGINEERING**

```python

age_bins = [0, 18, 25, 30, 40, 100]  
age_labels = ['Teenage', 'Young adult', 'Adult', 'Mid-age adult', 'Senior']  
binned_data = pd.cut(data['age_at_enrollment'], bins= age_bins, labels = age_labels)


data['life_stage'] = binned_data

```

```python
data['life_stage'] = data['life_stage'].astype('category')
```

***Aggregate Curricular Units: You have columns for Curricular units for both semesters. Let's create aggregate features such as total credited units, total approved units, pass rate, and total grades.***

```python
# Total credited units (sum of 1st and 2nd semesters)
data['Total credited units'] = data['curricular_units_1st_sem_credited'] + data['curricular_units_2nd_sem_credited']

# Total approved units (sum of approved units across both semesters)
data['Total approved units'] = data['curricular_units_1st_sem_approved'] + data['curricular_units_2nd_sem_approved']

# Pass rate (approved units/enrolled units)
data['Pass rate'] = (data['Total approved units'] / 
                   (data['curricular_units_1st_sem_enrolled'] + data['curricular_units_2nd_sem_enrolled'])).fillna(0)

# Total grades (average grade across all semesters)
data['Average grade'] = (data['curricular_units_1st_sem_grade'] + data['curricular_units_2nd_sem_grade']) / 2
```

## **LABEL ENCODING**

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

data['target'] = data['target'].astype('category')
data['life_stage'] = data['life_stage'].astype('category')

data['target_encoded'] = label_encoder.fit_transform(data['target'])

data['life_stage_encoded'] = label_encoder.fit_transform(data['life_stage'])

print(data[['target', 'target_encoded', 'life_stage', 'life_stage_encoded']].head())
```

## **HYPOTHESIS TEST**

***t-tests (Continuous Variables)***

```python
from scipy.stats import ttest_ind

dropouts = data[data['target'] == 'Dropout']['admission_grade']
non_dropouts = data[data['target'] != 'Dropout']['admission_grade']


t_stat, p_value = ttest_ind(dropouts, non_dropouts)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

***Insight: dropouts and non-dropouts differ significantly in terms of their admission grades, with dropouts typically having lower grades. The statistical significance of this difference is very high, suggesting that the relationship is unlikely to be due to random variation.***

***Chi-Square Test (Categorical Variables)***

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(data['gender'], data['target'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2}, P-value: {p}")
```

***Insight: there is a strong and statistically significant relationship between 'gender' and 'target', meaning that the distribution of the target variable significantly differs across genders.***
